```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data ={
    'Name':['Ravi','Priya','Aniket','Sara'],
    'Math':[78,92,67,85],
    'Science':[83,89,74,91],
    'English':[88,90,69,87]
}
df=pd.DataFrame(data)
df
```

```python
df['Total']= df[['Math','Science','English']].sum(axis=1)
df['Average']= df['Total']/3

def calculate_grade(avg):
    if avg >=90:
      return 'A+'
    elif avg >= 75:
      return 'A'
    elif avg >= 60:
      return 'B'
    elif avg >= 45:
      return 'C'
    else:
      return 'F'

df['Grade']= df['Average'].apply(calculate_grade)
df
```

```python
print("Subject-wise average marks:")
print(df[['Math','Science','English']].mean())

print("\nClass Toppers:")
print(df.loc[df['Total'].idxmax()])
```

```python
student_name = 'Priya'
student = df[df['Name'] == student_name].iloc[0]

subjects = ['Math', 'Science', 'English']
marks = [student['Math'], student['Science'], student['English']]

plt.figure(figsize=(6,4))
plt.bar(subjects, marks, color='skyblue')
plt.title(f'Marks of {student_name}')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.ylim(0, 100)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.bar(df['Name'], df['Average'], color='orange')
plt.title('Average Marks of Each Student')
plt.xlabel('Student Name')
plt.ylabel('Average Marks')
plt.ylim(0, 100)
plt.show()
```
